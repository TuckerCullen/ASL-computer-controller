### GroupBy gloss and video_Id

```python
df = df.sort_values('frame_id')
```

```python
df_list = df.groupby(['gloss', 'video_id']).agg(lambda x: str(list(x)))
```

```python
df_list = df_list.drop(['frame_id'], axis=1)
```

```python
df_list = df_list.reset_index()
```

### Label

```python
gloss = df_list['gloss'].unique()
```

```python
label_map = {}
idx = 0
for g in gloss:
    label_map[g] = idx
    idx = idx + 1
```

```python
df_list['gloss'] = df_list['gloss'].map(label_map)
```

```python
df_list = df_list.rename(columns={"gloss": "labels"})
```

```python
df_list.to_csv('wlasl2000_processedkp.csv', index=False)
```

### Train Valid Test Split

```python
from sklearn.model_selection import train_test_split
```

The ratio is 70%, 15%, 15%.

```python
def getSubClasses(df, num_classes, label_map):
    idx = random.sample(list(label_map.values()), num_classes)
    sub_df = df[df['labels'].isin(idx)]
    train, test = train_test_split(sub_df, test_size=0.3)
    valid, test = train_test_split(test, test_size=0.5)
    return train, valid, test
```

```python
subclasses = [100, 300, 500, 1000]
for i in subclasses:
    train, valid, test = getSubClasses(df_list, i, label_map)
    print("training value:",  len(train))
    print("training classes:", len(train['labels'].value_counts()))
    print("validation value:",  len(valid))
    print("validation classes:", len(valid['labels'].value_counts()))
    print("testing value:",  len(test))
    print("testing classes:", len(test['labels'].value_counts()))
    train.to_csv('train_' + str(i) + '.csv', index=False)
    valid.to_csv('valid_' +  str(i) + '.csv', index=False)
    test.to_csv('test_' +  str(i) + '.csv', index=False)
```
