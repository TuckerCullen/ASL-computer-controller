```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("keypoints.csv")
```

```python
# clean frame_id: remove .jpg & remove floats
def removeSuffix(x):
    if x.split(".")[-1] == "jpg":
        x = x.split(".")[0]
    return x

def frameidtoInt(x):
    return int(float(x) * 2)

df['frame_id'] = df['frame_id'].map(removeSuffix)
df['frame_id'] = df['frame_id'].map(frameidtoInt)
```

### Interpolate NaN

```python
# split dataframe by grouping
gf = df.groupby(['gloss', 'video_id'] )
```

```python
# for each group with frame number >= 70:
# 1) sort by frame_id
# 2) fill NaN backward, then forward
# 3) select 70 frames evenly
# 4) concate
def removeNan(grouped_dataframe):
    newdf = pd.DataFrame()
    for key in gf.groups.keys():
        subgf = gf.get_group(key)
        if subfg.size/45 >= 70:
            subgf = subgf.sort_values(by=['frame_id'])    # sorting

            subgf = subgf.fillna(method='bfill')          # fill in NaN
            subgf = subgf.fillna(method='ffill')

            frameIds = subgf.frame_id                                              # selection
            idx = np.round(np.linspace(0, len(frameIds) - 1, 70)).astype(int)
            sampleframeIds = np.array(frameIds)[idx]
            subgf = subgf[subgf['frame_id'].isin(sampleframeIds)]

            print("concatenating" + str(key))
            newdf = pd.concat([newdf, subgf])             # concatenation
    return newdf
```
